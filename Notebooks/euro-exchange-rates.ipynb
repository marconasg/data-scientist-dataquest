# Storytelling com Visualização de Dados sobre Taxas de Câmbio

## Introduzindo o dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

---

## Limpeza dos dados

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                                'Period\\Unit:': 'Time'},
                    inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

---

## Visualização dos dados

### Média móvel

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.title('Valores originais', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                            [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],
            euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Média móvel:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() # Ajusta automaticamente o espaçamento entre os gráficos
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

### Estabelecendo uma história pra contar

Algumas ideias de histórias para contar com nossos dados:
* Como a pandemia do coronavírus influenciou nas taxas euro-dólar. Podemos mostrar os dados de 2020 e comparar com 2016-2019, utilizando gráfico de linhas.
* Apresentar como a crise financeira de 2007-2008 afetou as taxas euro-dólar. Comparar com 2006 e 2009, para ter o antes e depois. Usar gráfico de linhas.
* Mostrar comparativamente como a taxa euro-dólar mudou sob os três últimos presidentes dos EUA (George W. Bush (2001-2009), Barack Obama (2009-2017) e Donald Trump (2017-2021)). Podemos usar um gráfico de linha.

#### Taxa de câmbio durante a pandemia do coronavírus

```python
covid = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2018) & 
                            (euro_to_dollar['Time'].dt.year <= 2021)]

after_covid = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2020) & 
                                    (euro_to_dollar['Time'].dt.year <= 2021)]
```

```python
# Adicionando estilo
import matplotlib.style as style
style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(11,6))
ax.plot(covid['Time'], covid['rolling_mean'], linewidth=1, color='#0000FF')
ax.plot(after_covid['Time'], after_covid['rolling_mean'], linewidth=3, color='#FF0000')

# Adicionando tick labels de forma separada
ax.set_xticklabels([])
ax.set_yticklabels([])

x = 17500.0
for year in ['2018', '2019', '2020', '2021']:
        ax.text(x, 1.07, year, alpha=0.5, fontsize=12)
        x += 370

y = 1.078   
for rate in ['1.08', '1.10', '1.12', '1.14', '1.16', '1.18', '1.20', '1.22', '1.24']:
        ax.text(17430.0, y, rate, alpha=0.5, fontsize=12)
        y += 0.02

# Realçando o pico durante a crise
ax.axvspan(xmin=18360.0, xmax=18425.0, ymin=0.018, ymax=0.5, alpha=0.5, color='grey')

# Título e subtítulo
ax.text(17430.0, 1.26, 'Taxa Euro-USD caiu para 1,085 durante a pandemia de coronavírus em 2020',
        weight='bold', size=20)
ax.text(17430.0, 1.25, 'Taxas de câmbio Euro-USD entre 2018 e 2021',
        size=16)

# Linhas de grades com opacidade
ax.grid(alpha=0.3)
```
